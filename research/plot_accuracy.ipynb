```python
import json

import numpy as np
import keras
from matplotlib import pyplot as plt
```

```python
with open('../models/efficientnet_v2_b3/orig/history_orig6.json') as f:
    history = json.load(f)

width = 0.3 
plt.figure(figsize=(10, 5))
plt.bar(np.arange(6) + 1 - width / 2, history['accuracy'], width, alpha=0.5)
plt.bar(np.arange(6) + 1 + width / 2, history['val_accuracy'], width, alpha=0.5)
plt.title('Точність класифікації для EfficientNetV2-B3', fontsize=16)
plt.ylabel('точність', fontsize=16)
plt.xlabel('номер епохи', fontsize=16)
plt.legend(['оригінальні тренувальні дані', 'оригінальні валідаційні дані'], loc='lower right', fontsize=12)
plt.ylim([0.975, 1.])
plt.savefig('orig_accuracy.png')
```

```python
with open('../models/efficientnet_v2_b3/syn_orig/history_syn6.json') as f:
    history1 = json.load(f)

with open('../models/efficientnet_v2_b3/syn_orig/history_orig3_after_syn6.json') as f:
    history2 = json.load(f)

plt.figure(figsize=(15, 5))
plt.bar(np.arange(6) + 1 - width / 2, history1['accuracy'], width, alpha=0.5, color='C2')
plt.bar(np.arange(6) + 1 + width / 2, history1['val_accuracy'], width, alpha=0.5, color='C3')
plt.bar(np.arange(3) + 7 - width / 2, history2['accuracy'], width, alpha=0.5, color='C0')
plt.bar(np.arange(3) + 7 + width / 2, history2['val_accuracy'], width, alpha=0.5, color='C1')
plt.title('Точність класифікації для EfficientNetV2-B3', fontsize=16)
plt.ylabel('точність', fontsize=16)
plt.xlabel('номер епохи', fontsize=16)
plt.legend(['синтетичні тренувальні дані',
            'синтетичні валідаційні дані',
            'оригінальні тренувальні дані',
            'оригінальні валідаційні дані'], loc='lower left', fontsize=16)
plt.ylim([0.965, 1.])
plt.savefig('syn_orig_accuracy.png')
```
